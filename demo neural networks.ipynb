# Neural network implementation
A neural network is a list of computation units.  A computation unit might have parameters.  The units transfrom input into output (feedforward).  But also they can update their parameters if passed an error (backpropagation).

```julia
abstract AbstractUnit               # an operation between layers, e.g. convolutional
# subtypes of AbstractUnit must implement:
#   sizeout(u, sizein)              # calculates the size of the output
#   feedforward(u, input)           # apply the unit to input and create output
#   backprop!(u,err,input,output)   # update the parameters and return the transformed error
#       note that 'err' has the same size as the output
```

# Linear unit

The forward processing is a matrix-vector multiplication:
$$ f(x) = W  x $$
where $W$ is an $m\times n$ matrix.
The error that gets passed is a $m$ dimensional vector $v$.  For the differential we get
$$ v^T d (Wx) = v^T W dx + tr v^T dW x = v^T W dx + \text{tr} x v^T dW $$
Thus the derivative of the loss function $\ell$ wrt to $W$ is:
$$ D_W \ell = x v^T $$
and the error that is propagated further is:
$$ D_x \ell = v^T W$$

```julia
type LinearUnit <: AbstractUnit            # fully connected operation
    W        # weight matrix
    lr       # learning rate
end
sizeout(u::LinearUnit, sizein) = sizein==(size(u.W,2),) ? (size(u.W,1),) : error("input size missmatch")
feedforward(u::LinearUnit, input) = u.W * input
backprop!(u::LinearUnit, derr, input, output) = begin
    Wtpderr = u.W' * derr                 # transform the error before the update
    u.W = u.W - u.lr * (derr * input')    # update the weights
    return Wtpderr                        # back propagate the transformed error
end
```

# Bias unit

```julia
type BiasUnit <: AbstractUnit            # fully connected operation
    b        # bias
    lr       # learning rate
end
sizeout(u::BiasUnit, sizein) = sizein
feedforward(u::BiasUnit, input) = input + u.b
backprop!(u::BiasUnit, derr, input, output) = begin
    u.b = u.b - u.lr * derr               # update the bias
    return derr                           # back propagate the transformed error
end
```

# Subsampling unit

```julia
type SubsamplingUnit <: AbstractUnit 
    strides
end
sizeout(u::SubsamplingUnit, sizein) = tuple([sizein[i]]...)
feedforward(u::SubsamplingUnit, input) = begin
    idx = [1:strides[i]:size(input,i) for i in 1:length(strides)]
    return input[idx...]
end
backprop!(u::SubsamplingUnint, derr, input, output) = begin
    derrout = zeros(size(input))
    idx = [1:strides[i]:size(input,i) for i in 1:length(strides)]
    derrout[idx] = derr
    return derrout
end
```

# Vectorizing unit

```julia
type VecUnit <: AbstractUnit end
sizeout(u::VecUnit, sizein) = (prod(sizein), )
feedforward(u::VecUnit, input) = vec(input)
backprop!(u::VecUnit, derr, input, output) = reshape(derr, size(input))
```

# Convolutional unit
Consider a matrix $x$ as input and a single filter $f$.  Then the output of the unit is the valid convolution of $x$ and $f$, possibly subsampled with some stride.

```julia
type ConvolutionUnit <: AbstractUnit
    filters  # array of filters, e.g. [5,5,10] for ten 5x5 filters
    lr       # learning rate
end
sizeout(u::ConvolutionUnit, sizein) = sizein>=size(squeeze(filters[:,:,1])) ? (size(u.W,1),) : error("input size missmatch")
feedforward(u::ConvolutionUnit, input) = begin
    0
end

backprop!(u::ConvolutionUnit, derr, input, output) = begin
    
    Wtpderr = u.W' * derr                 # transform the error before the update
    u.W = u.W - u.lr * (derr * input')    # update the weights
    return Wtpderr                        # back propagate the transformed error
end
```

# Sigmoidal unit
The forward processing is a calculating component-wise the sigmoidal function
$$ \sigma(t) = \frac{1}{1+e^{-t}} $$
The differential of the sigmoid has a special form:
$$ d \sigma(t) = \sigma(t)(1-\sigma(t)) dt $$
Note that the sigmoidal unit operates component-wise, i.e. for a vector $x$ we get,
$$ \sigma(x) = [\sigma(x_1), \ldots, \sigma(x_n)]$$
Similarly for the differential:
$$ d\sigma(x) = \text{Diag}(\sigma(x) \odot (1-\sigma(x))) dx$$
There are no parameters to update.  But the error $v$ has to be propagated correctly.  Consider a vector $x$
$$ v^T d\sigma(x) = v^T \text{Diag}(\sigma(x) \odot (1-\sigma(x))) dx$$

```julia
type SigmoidUnit <: AbstractUnit end
sizeout(u::SigmoidUnit, sizein) = sizein
feedforward(u::SigmoidUnit, input) = 1 ./ (1 .+ exp(-input))
backprop!(u::SigmoidUnit, derr, input, output) = derr .* (output.*(1-output))
```

# Rectified linear unit
The rectified linear unit calculates
$$ f(x) = \max(x, 0) $$
Thus their derivative is `diagm(x .>= 0)`, which is a matrix with ones and zeros along the diagonal.

```julia
type ReLUnit <: AbstractUnit end
sizeout(u::ReLUnit, sizein) = sizein
feedforward(u::ReLUnit, input) = max(input, 0)
backprop!(u::ReLUnit, derr, input, output) = derr .* float64(input .>= 0)
```

# tanh unit
The tangens hyperbolicus unit calculates
$$ f(x) = \tanh(x) $$
Thus their derivative is `diagm(x .>= 0)`, which is a matrix with ones and zeros along the diagonal.

```julia
type TanhUnit <: AbstractUnit end
sizeout(u::TanhUnit, sizein) = sizein
feedforward(u::TanhUnit, input) = tanh(input)
backprop!(u::TanhUnit, derr, input, output) = derr .* (1 - output.^2)
```

# Loss functions
For now we define only squared loss and its derivative:

```julia
loss(ytrue, y)  = 0.5 * vecnorm( ytrue - y )^2     # squared loss, should work for arbitrary arrays
dloss(ytrue, y) = ytrue - y                        # derivative of squared loss
```

# Neural networks
Neural networks are lists of units.  Since in our implementation the units do not store the values of the nodes, we also store the values in the neural networks.

```julia
type NeuralNetwork
    units::Array{AbstractUnit, 1}  # the list of units
    lunits                         # number of units
    sizes                          # the sizes of the layers
    values                         # the list of values at each layer
    function NeuralNetwork(units::Array{AbstractUnit, 1}, sizein)
        lunits = length(units)
        sizes  = Array(Any, lunits+1) # one longer to store input as well
        sizes[1] = sizein
        for i in 1:lunits
            sizes[i+1] = sizeout(units[i], sizes[i])
        end
        values = Array(Any, lunits+1)
        return new(units, lunits, sizes, values)
    end
end

function NeuralNetwork(arch::Array{Int64,1}; lr=1e-1, nonlinearUnit=SigmoidUnit)
    larch = length(arch)
    units = Array(AbstractUnit, 2*larch-3)
    j     = 1
    for i in 2:larch
        (m,n) = (arch[i], arch[i-1])           # output and input size
        W = sqrt(n) * randn(m, n)              # standard deviation sqrt(n)
        units[j] = LinearUnit(W, lr/n)         # learning rate normalized by input layer size
        j = j + 1
        if i < length(arch)
            units[j] = nonlinearUnit()
            j = j + 1
        end
    end
    return NeuralNetwork(units, (arch[1],))
end

function feedforward!(nn::NeuralNetwork, input)
    nn.values[1] = input
    for i in 1:length(nn.units)
        nn.values[i+1] = feedforward(nn.units[i], nn.values[i])
    end
    return nn.values[end]      # the output are the values at the last layer
end

function backprop!(nn::NeuralNetwork, input, output)
    ffoutput = feedforward!(nn, input)    # fills all layers
    derr = dloss(ffoutput, output)        # corresponds to the squared loss function
                                          # note that using -derr leads to divergence
    for i in nn.lunits:-1:1
        input  = nn.values[i]
        output = nn.values[i+1]
        derr   = backprop!(nn.units[i], derr, input, output)  
                                          # update the parameters of the units along the way
    end
    return ffoutput                       # the output of the feedforward
end

function train!(nn::NeuralNetwork, X, y)
    n = size(X,1)                # number of training examples
    errs = Array(Float64, n)     # the training errors
    for i in 1:n                 # stochastic gradient descent
        errs[i] = loss(y[i], backprop!(nn,vec(X[i,:]),y[i]))
    end
    return errs
end

apply!(nn::NeuralNetwork, X) = [feedforward!(nn, vec(X[i,:])) for i in 1:size(X,1)]

meanloss(nn, X, y) = mean([loss(ytrue, yest) for (ytrue, yest) in zip(y, apply!(nn, X))])
```

# Run on iris data set for debugging

```julia
using RDatasets
iris = dataset("datasets", "iris")         # using RDatasets
classes = unique(iris[5])
posclass = classes[2]    # PICK YOUR CLASS HERE
X      = array(iris[1:4])
C      = {"virginica"=>1e0, "setosa"=>1e0, "versicolor"=>1e0}[posclass]    # to store "good" values for C per class
y      = [label == posclass ? 1.0 : 0.0 for label in iris[5]]  # convert to +1.0 and 0.0
srand(17)        # set the random function to a fixed initial state for reproducibility
train  = randbool(size(iris, 1))       # create a random split in training and test set
Xtrain = X[ train, :]
ytrain = y[ train]
Xtest  = X[~train, :]
ytest  = y[~train]
```

```julia
nn = NeuralNetwork([4, 10, 1], lr=1e-1)     # define neural network
err = train!(nn, [Xtrain;Xtrain], [ytrain, ytrain])
plot(err)
```

```julia
nn = NeuralNetwork([4, 10, 1], lr=1e-1, nonlinearUnit=SigmoidUnit)     # define neural network
ntrain   = size(Xtrain, 1)           # number of training examples
ntest    = size(Xtest, 1)            # number of test examples
errtest  = zeros(ntrain)
errtrain = zeros(ntrain)
for i in 1:size(Xtrain, 1)
    train!(nn, Xtrain[i,:], y[i])
    errtest[i]  = meanloss(nn, Xtest, ytest)
    errtrain[i] = meanloss(nn, Xtrain, ytrain)
end
plot(x=[1:ntrain, 1:ntrain], y = [errtrain, errtest], color=[i<=ntrain ? "train" : "test" for i in 1:(2*ntrain)])
```
