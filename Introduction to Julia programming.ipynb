# How to define a function that works for scalars, vectors and matrices alike

Some scalar-valued should work for scalars and vectors.  How can we define that correctly?  Let's consider the dot product of two vectors $x'y$, aka inner product.  Note that for convenience this should also work for matrices and scalars.

```julia
dp(x::Vector, y::Vector) = x'*y
```

There are two literals which might be vectors, which to use?  Let's also try the literal for a matrix:

```julia
println(isa([1,2], Vector))        # USE THIS ONE
println(isa([1 2], Vector))        # DON'T USE THIS
println(isa([1 2; 3 4], Matrix))   # USE THIS ONE
```

So we should use `[1,2]` for `Vector`, and `[1 2; 3 4]` for `Matrix`.

```julia
dp(x,y) = x*y
```
