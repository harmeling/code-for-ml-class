# Amazon seller example (from Murphy's book)

Two sellers at Amazon have the same price. One has 90 positive, 10 negative reviews. The other one 2 positive, 0 negative. Who should you buy from?

Model the probability that the seller 1 and seller 2 receive a positive review as $\theta_1$ and $\theta_2$.  Assuming a beta-binomial model we obtain beta-distributed posteriors:
$$Beta(\theta_1|81,11)$$
$$Beta(\theta_2|3,1)$$

We are interested in the probability that seller 1 is more likely to receive a positive review than seller 2, i.e. we want to calculate:
$$p(\theta_1>\theta_2) = \int_0^1\int_0^1 [\theta_1>\theta_2] p(\theta_1|81,11) p(\theta_2|3,1) d\theta_1 d\theta_2$$

For this we sample from two beta distributions and calculate a ratio:

```julia
using Distributions   # if missing install with Pkg.add("Distributions")
```

```julia
n = 100000
theta1 = Beta(81,11)
theta2 = Beta(3,1)
sum(rand(theta1, n) .> rand(theta2, n)) / n   # wanted probability
```

```julia
using Winston
```

```julia
n = 100
theta = linspace(0, 1, n)
plot(theta, pdf(theta1, theta), theta, pdf(theta2, theta))
xlabel("theta")
ylabel("p(theta)")
```
